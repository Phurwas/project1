# 9.Adding new column

```python
new_col=[]
```

```python
for row in MAC270['Test_score']:
    if row > 95:
        new_col.append('A')
        
    elif row > 90:
        new_col.append('A-')
    elif row > 85:
        new_col.append('B')
    elif row > 80:
        new_col.append('B-')
    elif row > 75:
        new_col.append('C')
    elif row > 70:
        new_col.append('C-')
    elif row > 65:
        new_col.append('D')
    elif row > 60:
        new_col.append('D-')
    else:
        new_col.append('F')
        
MAC270['new_col']= new_col
```
